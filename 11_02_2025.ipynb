```python

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

mu = 100
sigma = 5
data = np.random.normal(mu, sigma, 1000)

plt.hist(data, bins=30, density=True, alpha=0.6, color='g')
plt.title('Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')

x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)

plt.plot(x, norm.pdf(x, mu, sigma), 'r', linewidth=2)
plt.show()

print("Mean:", np.mean(data))
print("Median:", np.median(data))
print("Mode (approximate, from the histogram):", data[np.argmax(np.histogram(data, bins=30)[0])])

from scipy.stats import skew

skewness = skew(data)
print("Skewness:", skewness)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats


mu, sigma = 0, 1
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
y = stats.norm.pdf(x, mu, sigma)

x1, x2 = -1, 1

plt.plot(x, y, 'k-')

plt.fill_between(x, y, where=(x >= x1) & (x <= x2), color='blue', alpha=0.5)

plt.fill_between(x, y, where=(x <= x1), color='red', alpha=0.5)

plt.fill_between(x, y, where=(x >= x2), color='green', alpha=0.5)

plt.title('Normal Distribution with Areas Under the Curve')
plt.xlabel('x')
plt.ylabel('Probability Density')

plt.show()
```

```python
import numpy as np
from scipy.stats import norm

def probability_between_iq_scores(iq1, iq2):
    mu = 100
    sigma = 15

    if not isinstance(iq1, (int, float)) or not isinstance(iq2, (int, float)):
        return "Error: Inputs must be numeric."
    if iq1 > iq2:
        return "Error: iq1 must be less than or equal to iq2."

    z1 = (iq1 - mu) / sigma
    z2 = (iq2 - mu) / sigma

    probability = norm.cdf(z2) - norm.cdf(z1)

    return probability

iq1 = float(input("Enter the first IQ score: "))
iq2 = float(input("Enter the second IQ score: "))

probability = probability_between_iq_scores(iq1, iq2)
probability

```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

mu = 100
sigma = 15

random_iq = np.random.normal(mu, sigma)

prob_greater_equal_85 = 1 - norm.cdf(85, loc=mu, scale=sigma)
prob_less_equal_85 = norm.cdf(85, loc=mu, scale=sigma)
prob_between_85_115 = norm.cdf(115, loc=mu, scale=sigma) - norm.cdf(85, loc=mu, scale=sigma)

categories = ['>= 85', '<= 85', '85-115']
probabilities = [prob_greater_equal_85, prob_less_equal_85, prob_between_85_115]

plt.figure(figsize=(8, 6))
plt.bar(categories, probabilities, color=['skyblue', 'lightcoral', 'lightgreen'])
plt.title('Probability Distribution of IQ Scores')
plt.xlabel('IQ Range')
plt.ylabel('Probability')

for i, prob in enumerate(probabilities):
    plt.text(i, prob + 0.01, f'{prob:.2f}', ha='center', va='bottom')

plt.ylim(0, max(probabilities) + 0.1)

plt.show()

print(f"Randomly selected IQ: {random_iq:.2f}")
print(f"Probability of IQ >= 85: {prob_greater_equal_85:.2f}")
print(f"Probability of IQ <= 85: {prob_less_equal_85:.2f}")
print(f"Probability of IQ between 85 and 115: {prob_between_85_115:.2f}")
```
